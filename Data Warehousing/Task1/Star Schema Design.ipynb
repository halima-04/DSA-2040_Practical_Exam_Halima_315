## 1. Star Schema Design

Fact Table: SalesFact
Measures:
TotalSales (calculated: Quantity × UnitPrice)
Quantity

Foreign Keys:
* CustomerID → CustomerDim
* ProductID → ProductDim
* TimeID → TimeDim

## Dimension Tables:

## CustomerDim
* CustomerID (PK)
* CustomerName (if available; otherwise null)
* Country

## ProductDim
* ProductID (PK)
* ProductName (Description)
* Category (you may categorize manually during ETL)
* StockCode

## TimeDim
* TimeID (PK)
* Date
* Month
* Quarter
* Year
